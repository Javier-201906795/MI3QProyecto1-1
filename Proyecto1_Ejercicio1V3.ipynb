```python
#Librerias
from sympy import * 
from numpy import *
from scipy.integrate import odeint
from matplotlib.pyplot import *
import matplotlib.pyplot as plt
from IPython.display import display
import numpy as np
from scipy.integrate import solve_ivp
#Simbolos
v,t = symbols ("v t")
```

```python
#Ecuacion diferencial
ED = 32-1.6*v
ED
```

```python
#Funcion de la Ecuacion Diferencial
def fED(v,t):
    return 32-1.6*v
```

```python
#Campo Direcicional
N=20 # numero de pendientes a graficar
xi=0; xf=400; # dominio
yi=0; yf=85; # rango
x=linspace(xi,xf,N);
y=linspace(yi,yf,N);
X,Y = meshgrid( x,y )
U = 1; V = fED(Y,X); U2, V2 = U/sqrt(U**2+V**2), V/sqrt(U**2+V**2)
quiver( X,Y,U2, V2,color='k')
grid(True)
title(r'Campo direccional: $ \frac{dP}{dt}=32-1.6*v $')
xlabel('Tiempo (t)')
ylabel('velocidad (v)')
```

```python
# Condicion inicial
v0 = 0
# Rango de integracion
t = np.linspace(0,10,700)
#Solucion de la Ecuacion Diferencial
sol = odeint(fED,v0,t)
# Crear la gráfica
plt.plot(t, sol)
plt.title(r'Curva Solucion de la ED: $ \frac{dv}{dt}=32-1.6v $')
plt.xlabel('Tiempo (t)')
plt.ylabel('Velocidad (v)')
plt.grid(True)
```

```python
# a.) Cual sera su velocidad Limite
def model(t, v):
    return 32 - 1.6 * v

# Definir el tiempo inicial y final
t_span = (0, 10)

# Definir la condición inicial
v0 = 0

# Resolver la ecuación diferencial
sol = solve_ivp(model, t_span, [v0], t_eval=np.linspace(0, 10, 100))

# Graficar la solución
plt.plot(sol.t, sol.y[0])
plt.xlabel('Tiempo')
plt.ylabel('v(t)')
plt.title('Solución de la Ecuación Diferencial')
plt.grid(True)
plt.show()

# Calcular el límite de la solución cuando t tiende a infinito
limite_v = sol.y[0][-1]
print("la velocidad límite de la solución es:", limite_v,'m/s')
```

```python
#c.)  Encontrar el tiempo en el que la velocidad alcanza el 95% de la velocidad límite
v_95 = 0.95 * limite_v

# Encontrar el tiempo en el que la velocidad alcanza el 95% de la velocidad límite
tiempo_95 = sol.t[np.argmax(sol.y[0] >= v_95)]

print("Tiempo para alcanzar el 95% de la velocidad límite:", tiempo_95, "segundos")
```
