```python
#Librerias
from sympy import *
from numpy import *
from scipy.integrate import odeint
from matplotlib.pyplot import *
import matplotlib.pyplot as plt
from IPython.display import display
#Simbolos
P,t = symbols("P t")
```

```python
#Ecucaion Diferencial
ED = 0.0225*P-0.0003*P**2
ED
```

```python
# Funcion de la Ecuacion Diferencial
def fED(P,t):
    return 0.0225 * P - 0.0003 * P**2
```

```python
#Campo Direcicional
N=20 # numero de pendientes a graficar
xi=0; xf=400; # dominio
yi=0; yf=85; # rango
x=linspace(xi,xf,N);
y=linspace(yi,yf,N);
X,Y = meshgrid( x,y )
U = 1; V = fED(Y,X); U2, V2 = U/sqrt(U**2+V**2), V/sqrt(U**2+V**2)
quiver( X,Y,U2, V2,color='k')
grid(True)
title(r'Campo direccional: $ \frac{dP}{dt}=0.0225P-0.0003P^2 $')
xlabel('Tiempo (t)')
ylabel('Poblacion (P)')
```

```python
# Condicion inicial
P0 = 25
# Rango de integracion
t = np.linspace(0,400,700)
#Solucion de la Ecuacion Diferencial
sol = odeint(fED,P0,t)
# Crear la grfica
plt.plot(t, sol)
plt.title(r'Curva Solucion de la ED: $ \frac{dP}{dt}=0.0225P-0.0003P^2 $')
plt.xlabel('Tiempo (t)')
plt.ylabel('Poblacion (P)')
plt.grid(True)
```

```python
#Cuanto tiempo le tomara duplicarse a esta poblacion?

# Encontrar el ndice donde P es aproximadamente igual a 50
index_P_50 = np.argmin(np.abs(sol - 50))

# Obtener el valor de t cuando P es igual a 50
tiempo_cuando_P_50 = t[index_P_50]
display('Cuanto tiempo le tomara duplicarse a esta poblacion?')
display(f'Se duplicara en t = {tiempo_cuando_P_50}')
```

```python
#Cual sera la poblacion limite de venados?
limite_infinito = sol[-1][0]
limite_infinito
display('Cual sera la poblacion limite de venados?')
display(f'La poblacin lmite de venados es: {limite_infinito}')
```

```python
# Crear la grfica
plt.plot(t, sol)
plt.title(r'Problema 2')
plt.xlabel('Tiempo (t)')
plt.ylabel('Poblacion (P)')
plt.grid(True)
# Agregar el punto
plt.scatter(tiempo_cuando_P_50, 50, color='red', marker='o', label='Doble P0', s=100)
# Agregar etiqueta al punto
plt.text(tiempo_cuando_P_50+14, 51, f'Duplicacion Poblacion inicial \n({round(tiempo_cuando_P_50, 3)},50)', fontsize=9, ha='left')
#Campo Direcicional
N=15 # numero de pendientes a graficar
xi=0; xf=400; # dominio
yi=0; yf=85; # rango
x=linspace(xi,xf,N);
y=linspace(yi,yf,N);
X,Y = meshgrid( x,y )
U = 1; V = fED(Y,X); U2, V2 = U/sqrt(U**2+V**2), V/sqrt(U**2+V**2)
quiver( X,Y,U2, V2,color='k')
grid(True)
```
